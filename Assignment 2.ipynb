```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np

from math import sin, pi
theta = np.linspace(-np.pi,40*np.pi, 100000)
r1= 1/(1.05409*10**(-6)*np.sin(.948683*theta)-9.11111*np.cos(0.948683*theta)+11.1111)
r2= 1/(5*(theta**2)+(theta*10**(-6))+2)


ax=plt.subplot(111,polar=True)
ax.plot(theta,r1,color='purple',marker='')
plt.show()
ax.plot(theta,r2,color='green',marker='')
plt.show()
```
